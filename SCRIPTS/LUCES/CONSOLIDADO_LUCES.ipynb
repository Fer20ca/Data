```python
os.chdir('C:/Users/FERNANDO/Documents/PI_INEQUIDAD/scripts_data/data/outputs/LUMINOSIDAD_PERU/LUCES_DISTRITOS')
```

```python
d=pd.read_csv('LucesNocturnas_Promedio_Distrital_2007.csv') 
```

```python
d.columns
```

```python
d_1=d.loc[:,['DEPARTAMEN','CAPITAL','PROVINCIA','DISTRITO','IDDIST','mean']].copy()
d_1=d_1.rename(columns={'mean': '2007'})
d_1
```

```python
d_consol=d_1.copy()
```

```python
archivo='LucesNocturnas_Promedio_Distrital_'

anio=list(range(2008,2024))

for elem in anio:
    print(elem)
    fecha=str(elem)
    archivo = f'LucesNocturnas_Promedio_Distrital_{fecha}.csv'
    # asignamos la data a iterar
    d = pd.read_csv(archivo)
    # a partir del año 2008 solo necesito sus valores por año de intensidad luminosa
    d_1 = d.loc[:,['IDDIST','mean']].copy()
    d_1 = d_1.rename(columns={'mean': fecha})
    d_consol = pd.merge(d_consol, d_1, on = "IDDIST", how='inner')  

d_consol
```

```python
d_consol.to_excel("C:/Users/FERNANDO/Documents/PI_INEQUIDAD/scripts_data/data/outputs/LUMINOSIDAD_PERU/CONSOLIDADOS/lum_distr_09_023.xlsx")
```
