# POLINÓMICA

```python
# --- 1. Asegurar que todas las columnas sean strings
data.columns = data.columns.astype(str)

# --- 2. Separar las columnas de años y las de identificación
year_cols = [c for c in data.columns if c.isdigit()]
id_cols = [c for c in data.columns if not c.isdigit()]

# --- 3. Ordenar los años numéricamente
year_cols = sorted(year_cols, key=int)

# --- 4. Reordenar el DataFrame
data = data[id_cols + year_cols]

# --- 5. Interpolación numérica entre años
# Convertimos los años a columnas numéricas para interpolar correctamente
data[year_cols] = data[year_cols].apply(pd.to_numeric, errors='coerce')


```

```python
# Realizamos la interpolación a lo largo de las columnas (por fila)
inter_lineal= data[year_cols].interpolate(axis=1, method='linear')

# --- 6. Verificamos los resultados
inter_lineal
```

```python
import pandas as pd

# Copiamos solo las columnas de años
temp = data[year_cols].copy()

# Convertimos los nombres de columnas a enteros
temp.columns = temp.columns.astype(int)

# Interpolamos con método polinómico de orden 2
inter_poli = temp.interpolate(axis=1, method='polynomial', order=2)

# Restauramos nombres originales (como string)
inter_poli.columns = inter_poli.columns.astype(str)

# Reemplazamos los valores en el DataFrame original
inter_poli
```

```python
temp = data[year_cols].copy()
temp.columns = pd.to_datetime(temp.columns, format='%Y')

# Interpolación basada en tiempo
inter_time = temp.interpolate(method='time', axis=1)

# Restauramos los nombres originales (como strings)
inter_time.columns = inter_time.columns.year.astype(str)

# Reemplazamos los valores interpolados en el DataFrame original
inter_time

```
