```python
data=pd.read_excel('7213200000 - 2015.xls')
data
```

```python
cols_a_mantener = [i for i in range(data.shape[1]) if i not in [0,3,4,6,8,13,12]]
data = data.iloc[:, cols_a_mantener]
data
```

```python
data=data.dropna()
```

```python
data
```

```python
data.columns = data.iloc[0]      # Toma la primera fila como encabezado
data = data[1:]                  # Elimina esa primera fila del cuerpo del DataFrame
df = data.reset_index(drop=True)  # Opcional: resetea el índice
df
```

```python
#df.dtypes# Lista de columnas a convertir
columnas_numericas = ['FOB $', 'CIF $', 'ADV $', 'IMP. ARANCEL $']

# Reemplazar comas y convertir a float
for col in columnas_numericas:
    df[col] = df[col].replace(',', '', regex=True).astype(float)

df.dtypes
```

```python
importaciones_por_pais = df.groupby('PAIS')['CIF $'].sum().sort_values(ascending=False)
print(importaciones_por_pais)
```

```python

df_2019 = importaciones_por_pais.to_frame(name='2015')
df_2019
df_total=df_2019.copy()
```

```python
df_total
```

```python
años=['2016','2017','2018','2019','2020','2021','2022','2023','2024']


for año in años:
    data_1=pd.read_excel(f'7213200000 - {año}.xls')
    cols_a_mantener = [i for i in range(data_1.shape[1]) if i not in [0,3,4,6,8,13,12]]
    data_1 = data_1.iloc[:, cols_a_mantener]
    data_1=data_1.dropna()
    data_1.columns = data_1.iloc[0]      # Toma la primera fila como encabezado
    data_1 = data_1[1:]                  # Elimina esa primera fila del cuerpo del DataFrame
    df_1 = data_1.reset_index(drop=True)  # Opcional: resetea el índice 
    #df.dtypes# Lista de columnas a convertir
    columnas_numericas = ['FOB $', 'CIF $', 'ADV $', 'IMP. ARANCEL $']

# Reemplazar comas y convertir a float
    for col in columnas_numericas:
        df_1[col] = df_1[col].replace(',', '', regex=True).astype(float)

    producto = df_1.groupby('PAIS')['CIF $'].sum().sort_values(ascending=False)
    
    df_app = producto.to_frame(name=f'{año}')

    df_total = df_total.merge(df_app, on='PAIS', how='outer')

df_total
```

```python
df_total.to_excel('total_7213200000.xlsx')
```
