```python
#INGRESO A LA WEB DEL INEI MICRODATOS EXACTAMENTE AL APARTADO ENCUESTAS
#https://datosabiertos.gob.pe/dataset/registro-nacional-de-municipalidades-renamu-2007-instituto-nacional-de-estad%C3%ADstica-e
# EL METODO PARA SCRAPEAR VA A SER DISTINTO
# SUCEDE QUE ESTA VEZ EN EL INEI CON SUS BASES DE ENCUESTAS SOLO TENEMOS LAS RENAMUS EN OTRO FORMATO
# LO QUE QUEREMOS ES EL CSV DISPONIBLE EN LA WEB DEL INE
# EN LA LINEA 2 DE ESTA CELDA SE MUESTRA EL FORMATO DE LA DIRECCIÓN DE CADA UNA DE LAS ENCUESTAS
```

```python

for year in range(2007, 2024):

    driver = webdriver.Chrome()
    driver.maximize_window()# sin ninguna acción maximiza
    url = f'https://datosabiertos.gob.pe/dataset/registro-nacional-de-municipalidades-renamu-{year}-instituto-nacional-de-estad%C3%ADstica-e'
    driver.get(url)

    time.sleep(2)

    # LE HACEMOS CLICK AL ELEMNTITO QUE TIENE FORMA DE ZIP .... XPATH: //*[@id="data-and-resources"]/div/div/ul/li[3]/div/a/span
    ingreso = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="data-and-resources"]/div/div/ul/li[3]/div/a/span'))
    )

    ingreso.click()

    time.sleep(2)

    # le damos click  a "IR AL RECURSO"
    descarga = WebDriverWait(driver, 20).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="main"]/div/section/ul/li[3]/a'))
    )

    descarga.click()
    # hare que el sistema se duerma unos 5  segundos para que se realice correctamrnte la descarga
    time.sleep(5)

    driver.close()

    time.sleep(2)
```
