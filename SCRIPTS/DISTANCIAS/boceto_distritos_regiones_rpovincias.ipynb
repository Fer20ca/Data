```python
import geopandas as gpd
import os 
import numpy as np
import pandas as pd
from geopy.geocoders import Nominatim
from unidecode import unidecode
```

```python
os.chdir('C:/Users/FERNANDO/Documents/PI_INEQUIDAD/scripts_data/data/outputs/distancias/cap_regionalxprovincias')
df=pd.read_excel('lon_lat_cap_prov.xlsx')
df[df.isna().any(axis=1)]
```

```python
import re

def limpiar_nombre(nombre):
    if pd.isna(nombre):
        return nombre
    nombre = re.sub(r"/\d+", "", nombre)  # elimina /4, /7, etc.
    nombre = re.sub(r"\(.*?\)", "", nombre)  # elimina (PUEBLO NUEVO DE COLAN)
    return nombre.strip()

df["capital_limpia"] = df["CAPITAL"].apply(limpiar_nombre)
```

```python
df_1=df[df.isna().any(axis=1)].copy()
```

```python
lat_long = [
    (-14.25006, -72.68175),
    (-14.9969, -73.3514),
    (-13.8522, -73.87729),
    (-15.0089, -71.6432),
    (-16.27001, -69.17745),
    (-16.565834, -71.368339),  
    (-12.87198, -72.44732),
    (-13.39675, -73.59358),
    (-7.3277, -78.8634),
    (-9.94367, -78.22478),
    (-6.45315, -77.45533),
    (-3.4901, -74.78112),
    (-5.006807, -81.05903)
]

# Crear un DataFrame con las coordenadas
df_1=df_1.reset_index(drop=True)
df_1[['lat', 'lon']] = pd.DataFrame(lat_long, columns=['lat', 'lon'])
```

```python
# Calcular centroides


gdf_final = gpd.GeoDataFrame(
    df_total,
    geometry=gpd.points_from_xy(df_total['lon'], df_total['lat']),
    crs="EPSG:4326"  # Sistema de coordenadas WGS84
)
```

```python
base_cap_distrital=gdf_final.iloc[:,[0,1,2,3,8]].copy()
```

```python

base_cap_distrital.rename(columns={'geometry': 'cor_cap_distr'}, inplace=True)
base_cap_distrital.rename(columns={'CAPITAL': 'cap_distr'}, inplace=True)

base_cap_distrital_sorted=base_cap_distrital.sort_values(by= ['NOMBDEP','NOMBPROV','NOMBDIST']).copy()
base_cap_distrital_sorted

base_cap_distrital_sorted['NOMBDEP'] = base_cap_distrital_sorted['NOMBDEP'].apply(lambda x: unidecode(x.strip().upper()))
base_cap_distrital_sorted['NOMBPROV'] = base_cap_distrital_sorted['NOMBPROV'].apply(lambda x: unidecode(x.strip().upper()))

#PRIMERO QUE EN ESTA BASE ESTA ESCRITO NASCA EN VEZ DE NAZCA
base_cap_distrital_sorted['NOMBPROV'] = base_cap_distrital_sorted['NOMBPROV'].replace('NASCA', 'NAZCA')
#SEGUNDO CONTRALMIRANTE VILLAR EN VEZ DE CONTRALMIRANTE VILLA
base_cap_distrital_sorted['NOMBPROV'] = base_cap_distrital_sorted['NOMBPROV'].replace('CONTRALMIRANTE VILLAR', 'CONTRALMIRANTE VILLA')
#TERCERO SAN ANTONIO DE PUTINA EN VEZ DE SAN ANTONIO DE PUTIN
base_cap_distrital_sorted['NOMBPROV'] = base_cap_distrital_sorted['NOMBPROV'].replace('SAN ANTONIO DE PUTINA', 'SAN ANTONIO DE PUTIN')
#CUARTO GENERAL SANCHEZ CERRO EN VEZ DE GRAL.SÁNCHEZ CERRO
base_cap_distrital_sorted['NOMBPROV'] = base_cap_distrital_sorted['NOMBPROV'].replace('GENERAL SANCHEZ CERRO', 'GRAL.SANCHEZ CERRO')
#QUINTO DANIEL ALCIDES CARRION EN VEZ DE DANIEL ALCIDES CARRI
base_cap_distrital_sorted['NOMBPROV'] = base_cap_distrital_sorted['NOMBPROV'].replace('DANIEL ALCIDES CARRION', 'DANIEL ALCIDES CARRI')
#SEXTO MARISCAL RAMÓN CASTILLA EN VEZ DE MARISCAL RAMÓN CASTI
base_cap_distrital_sorted['NOMBPROV'] = base_cap_distrital_sorted['NOMBPROV'].replace('MARISCAL RAMON CASTILLA', 'MARISCAL RAMON CASTI')
#SEPTIMO CARLOS FERMIN FITZCARRALD EN VEZ DE CARLOS F. FITZCARRAL
base_cap_distrital_sorted['NOMBPROV'] = base_cap_distrital_sorted['NOMBPROV'].replace('CARLOS FERMIN FITZCARRALD', 'CARLOS F. FITZCARRAL')
#OOCTAVO FERRENAFE
base_cap_distrital_sorted['NOMBPROV'] = base_cap_distrital_sorted['NOMBPROV'].replace('CARLOS FERMIN FITZCARRALD', 'CARLOS F. FITZCARRAL')
```

```python
base_prov_reg=pd.read_excel('dist_provincia_cap_to_cap_region.xlsx')
base_prov_reg.rename(columns={'capital': 'cap_reg'}, inplace=True)
base_prov_reg.rename(columns={'Nombre': 'cap_prov'}, inplace=True)
base_prov_reg.rename(columns={'Provincia': 'provincia'}, inplace=True)
base_prov_reg

base_prov_reg['region'] = base_prov_reg['region'].apply(lambda x: unidecode(x.strip().upper()))
base_prov_reg['provincia'] = base_prov_reg['provincia'].apply(lambda x: unidecode(x.strip().upper()))
```

```python
# al hacer el merge deberia de continuar con los 1891 registros de municipalidades
merged = base_cap_distrital_sorted.merge(base_prov_reg, left_on=['NOMBDEP','NOMBPROV'], right_on=['region','provincia'], how='left')
```

```python
merged
```

```python
merged = merged.set_geometry("cor_cap_distr")

# 1️⃣ Definir CRS original (solo una vez, no transforma coordenadas)
merged = merged.set_crs("EPSG:4326")

merged=merged.to_crs("EPSG:32718").copy()
merged
```

```python
df_final=merged.loc[:,['NOMBDEP', 'NOMBPROV', 'NOMBDIST', 'cap_reg', 'cap_prov', 'cap_distr', 'cap_distr', 'cap_region', 'cap_provincia', 'cor_cap_distr']].copy()
```

```python
from shapely import wkt

def convertir_wkt_si_es_texto(col):
    return col.apply(lambda x: wkt.loads(x) if isinstance(x, str) and x.strip().startswith("POINT") else x)

df_final['cap_provincia'] = convertir_wkt_si_es_texto(df_final['cap_provincia'])
df_final['cap_region'] = convertir_wkt_si_es_texto(df_final['cap_region'])
df_final['cor_cap_distr'] = convertir_wkt_si_es_texto(df_final['cor_cap_distr'])
df_final = df_final.set_geometry("cor_cap_distr")
df_final = df_final.set_geometry("cap_region")
df_final = df_final.set_geometry("cap_provincia")
df_final
```
