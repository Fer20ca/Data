```python
import os
import requests
from datetime import datetime
import re
import json  # Para exportar el resumen
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import numpy as np
```

```python
inicio="1950" # inicio
fin="2021" # termina en 2024 todas las series 
fechas = pd.date_range(start=inicio, end=fin, freq='YS') # creamos las fechas para que tengan solo mes y año
fechas_lista = fechas.strftime('%Y').tolist() # ACA ESTAN LAS FECHAS
print(fechas_lista)
```

```python
# ESTA PRIMERA CELDA ES UNA MUESTRA DE LO QUE SE HACE
codigo="PD39589DA" 
url = f'https://estadisticas.bcrp.gob.pe/estadisticas/series/api/{codigo}/json/1950/2021' # las fechas de inicio y final tmbn pueden ser delimitadas
response = requests.get(url) # la respuesta viene a ser el output de la consulta de la dirección dada por el API
data_publica=response.json() # lo almacenamos en un elemenot llamado data
print(data_publica) # la data es un diccionario que contiene todos los valores que necesitamos
```

```python
# ESTA PRIMERA CELDA ES UNA MUESTRA DE LO QUE SE HACE
codigo="PD39588DA" 
url = f'https://estadisticas.bcrp.gob.pe/estadisticas/series/api/{codigo}/json/1950/2021' # las fechas de inicio y final tmbn pueden ser delimitadas
response = requests.get(url) # la respuesta viene a ser el output de la consulta de la dirección dada por el API
data_privada=response.json() # lo almacenamos en un elemenot llamado data
print(data_privada) # la data es un diccionario que contiene todos los valores que necesitamos
```

```python
datos_publica=[float(elem['values'][0]) for elem in data_publica['periods']]
print(datos_publica)

datos_privada=[float(elem['values'][0]) for elem in data_privada['periods']]
print(datos_privada)

```

```python
columnas = ['Fecha','INV_publica','INV_privada']
datos= [fechas_lista]+[datos_publica]+[datos_privada]
dic_data={}

for i in range(len(datos)): # se arma el diccionario con la data correspondiente a cada columna
    dic_data[columnas[i]]=datos[i]


df=pd.DataFrame(dic_data,columns=columnas) # ya estan en unidades reales (NO CONFUNDIR)
df = df.apply(pd.to_numeric, errors='coerce')
df
```
