```python
anios=['2006','2010','2014','2018','2022']
dfs_multindex=[]
for anio in anios:

    dir_base = fr"C:\Users\FERNANDO\Documents\PI_INEQUIDAD\scripts_data\data\outputs\elecciones\municipalidades_distritales\MUNICIPAL DISTRITAL {anio}"
    ruta_archivo = fr"{dir_base}\ERM{anio}_Resultados_Distrital.xlsx"
    data = pd.read_excel(ruta_archivo)


    
    try:
        conteo = data["Distrito"].value_counts().reset_index()
        conteo.columns = ["Distrito", "cantidad_postulantes"]
        conteo=conteo.sort_values(by='Distrito').reset_index(drop=True)
        print('listo')

        # tabla con el porcentaje de ganadores
        ganadores = data.loc[data.groupby("Distrito")["% Votos"].idxmax()]
        ganadores
        # de acá nos quedaremos con que tipo de partido político fueron el de los ganadores
        ganadores=ganadores.loc[:,['Región','Provincia','Distrito','Tipo Organización Política','% Votos','Votos']]
        ganadores=ganadores.reset_index(drop=True)

        print('listo')

        # ahora debo de juntar estos
        merged = pd.merge(conteo, ganadores, on="Distrito")
        merged=merged.sort_values(by='Región').copy()
        nuevo_orden = ['Región', 'Provincia','Distrito', 'Tipo Organización Política', '% Votos', 'Votos', 'cantidad_postulantes']
        merged = merged[nuevo_orden].reset_index(drop=True)
        dfs_multindex.append(merged)
        print('listo')


    except KeyError:
        conteo = data["Distrito"].value_counts().reset_index()
        conteo.columns = ["Distrito", "cantidad_postulantes"]
        conteo=conteo.sort_values(by='Distrito').reset_index(drop=True)
        print('listo')
        # tabla con el porcentaje de ganadores
        # tabla con el porcentaje de ganadores
        ganadores = data.loc[data.groupby("Distrito")["% Votos"].idxmax()]
        # de acá nos quedaremos con que tipo de partido político fueron el de los ganadores
        ganadores=ganadores.loc[:,['Region','Provincia','Distrito','Tipo Organización Política','% Votos','Votos']]
        ganadores=ganadores.reset_index(drop=True)
        print('listo')
        # ahora debo de juntar estos
        merged = pd.merge(conteo, ganadores, on="Distrito")
        merged=merged.sort_values(by='Region').copy()
        nuevo_orden = ['Region', 'Provincia','Distrito', 'Tipo Organización Política', '% Votos', 'Votos', 'cantidad_postulantes']
        merged = merged[nuevo_orden].reset_index(drop=True)
        merged = merged.rename(columns={"Region": "Región"})
        dfs_multindex.append(merged)
        print('listo')


print(len(dfs_multindex)) 
```

```python
# Nos aseguramos de que todos tengan 'Región', 'Provincia' y 'Distrito' como índice
for i, df in enumerate(dfs_multindex):
    dfs_multindex[i] = df.set_index(["Región", "Provincia", "Distrito"])

# Concatenamos horizontalmente con jerarquía en columnas
data_multi = pd.concat(dfs_multindex, axis=1, keys=anios)

# Reordenamos los niveles: 1er nivel = variables, 2do = años
data_multi = data_multi.swaplevel(axis=1).sort_index(axis=1)

data_multi.head()
```
