```python
anios=['2006','2010','2014','2018','2022']
dfs_multindex=[]
for anio in anios:
    print(anio)
    dir_base = fr"C:\Users\FERNANDO\Documents\PI_INEQUIDAD\scripts_data\data\outputs\elecciones\regionales\REGIONAL {anio}"
    ruta_archivo = fr"{dir_base}\ERM{anio}_Resultados_Regional.xlsx"
    data = pd.read_excel(ruta_archivo)
    
    try:
        conteo = data["Región"].value_counts().reset_index()
        conteo.columns = ["Región", "cantidad_postulantes"]
        conteo=conteo.sort_values(by='Región').reset_index(drop=True)

        # tabla con el porcentaje de ganadores
        ganadores = data.loc[data.groupby("Región")["% Votos"].idxmax()]

        # de acá nos quedaremos con que tipo de partido político fueron el de los ganadores
        ganadores=ganadores.loc[:,['Región','Tipo Organización Política','% Votos','Votos']]
        ganadores=ganadores.reset_index(drop=True)

        # ahora debo de juntar estos
        merged = pd.merge(conteo, ganadores, on="Región")
        dfs_multindex.append(merged)

    except KeyError:
        conteo = data["Region"].value_counts().reset_index()
        conteo.columns = ["Region", "cantidad_postulantes"]
        conteo=conteo.sort_values(by='Region').reset_index(drop=True)

        # tabla con el porcentaje de ganadores
        ganadores = data.loc[data.groupby("Region")["% Votos"].idxmax()]

        # de acá nos quedaremos con que tipo de partido político fueron el de los ganadores
        ganadores=ganadores.loc[:,['Region','Tipo Organización Política','% Votos','Votos']]
        ganadores=ganadores.reset_index(drop=True)

        # ahora debo de juntar estos
        merged = pd.merge(conteo, ganadores, on="Region")
        merged = merged.rename(columns={"Region": "Región"})
        dfs_multindex.append(merged)

print(len(dfs_multindex)) 
```

```python
# Nos aseguramos de que todos tengan 'Región' como índice
for i, df in enumerate(dfs_multindex):
    dfs_multindex[i] = df.set_index("Región")

# Concatenamos horizontalmente con jerarquía en columnas
data_multi = pd.concat(dfs_multindex, axis=1, keys=anios)

# Reordenamos los niveles para que el 1er nivel sean las variables y el 2do los años
data_multi = data_multi.swaplevel(axis=1).sort_index(axis=1)

data_multi.head()
```
